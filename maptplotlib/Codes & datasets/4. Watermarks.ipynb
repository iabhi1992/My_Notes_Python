```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

* Plotting a lines and using x and y values

```python
fig, ax = plt.subplots()

ax.plot([3, 5, 6 , 8, 10, 11], 
         [1, 4, 7, 5, 11, 10],'-.o',
        mfc='#FF5733',markersize=10)
```

* Adding text annotation, and it defines the position of the text and the horizondal and vertical algnment positions

```python
fig, ax = plt.subplots()

ax.plot([3, 5, 6 , 8, 10, 11], 
         [1, 4, 7, 5, 11, 10],'-.o',
        mfc='#FF5733',markersize=10)

ax.text(4, 9, 'Save Nature',
        fontsize=30, 
        color='green',
        ha='left', 
        va='bottom', 
        alpha=0.5)
```

* Changing the position of text annotation

```python
fig, ax = plt.subplots()

ax.plot([3, 5, 6 , 8, 10, 11], 
         [1, 4, 7, 5, 11, 10],'-.o',
        mfc='#FF5733',markersize=10)

ax.text(4, 9, 'Save Nature',
        fontsize=30, 
        color='green',
        ha='right', 
        va='bottom', 
        alpha=0.5)
```

```python
fig, ax = plt.subplots()

ax.plot([3, 5, 6 , 8, 10, 11], 
         [1, 4, 7, 5, 11, 10],'-.o',
        mfc='#FF5733',markersize=10)

ax.text(10, 9, 'Save Nature',
        fontsize=30, 
        color='green',
        ha='left', 
        va='top', 
        alpha=0.5)
```

```python
x = np.linspace(start=1, stop=15, num=60)
```

* Changing the figure size
* And arranging the text inside of the plot

```python
fig = plt.figure(figsize=(6,6)) 

ax1 = fig.add_subplot(221)
                      
ax1.plot([3, 5, 6 , 8, 10, 11], 
         [1, 4, 7, 5, 11, 10],'-.o',
        mfc='#FF5733',markersize=10)
ax1.set_xlabel('lines')

ax1.text(3, 9, 'Save Nature',
        fontsize=20, 
        color='green',
        ha='left', 
        va='top', 
        alpha=0.5)
```

* Adding multiple plots and multiple text annotations

```python
fig = plt.figure(figsize=(6,6))

ax1 = fig.add_subplot(221)
ax1.plot([3, 5, 6 , 8, 10, 11], 
         [1, 4, 7, 5, 11, 10],'-.o',
        mfc='#FF5733',markersize=10)
ax1.set_xlabel('lines')

ax1.text(3, 9, 'Save Nature',
        fontsize=20, 
        color='green',
        ha='left', 
        va='top', 
        alpha=0.5)

ax2 = fig.add_subplot(222)
ax2.plot(x, np.trunc(x))

ax3 = fig.add_subplot(223)
ax3.plot(x, np.tan(x))

ax3.text(8, 20, 'Save Water',
        fontsize=20, 
        color='#00CED1',
        ha='center', 
        va='top', 
        alpha=0.5)

plt.show
```
