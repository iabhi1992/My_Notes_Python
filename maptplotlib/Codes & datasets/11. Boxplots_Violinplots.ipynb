```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
x = np.random.randint(low=0, high=30, size=30)
x.sort()
x
```

```python
plt.boxplot(x)

plt.show()
```

```python
x = np.append(x, 22)

plt.boxplot(x) 

plt.show()
```

```python
x = np.append(x, 37)
x = np.append(x, 40)

plt.boxplot(x) 

plt.show()
```

```python
plt.boxplot(x, vert=False) 

plt.show()
```

```python
plt.boxplot(x, 
            vert=False,
            notch=True) 

plt.show()
```

```python
plt.boxplot(x, 
            vert=False,
            notch=True,
            showfliers=False) 

plt.show()
```

```python
bp = plt.boxplot(x, patch_artist=True) 
bp
```

```python
print('Number of boxes: ', len(bp['boxes']))
print('Number of whiskers: ', len(bp['whiskers']))
print('Number of caps: ', len(bp['caps']))
print('Number of medians: ', len(bp['medians']))
print('Number of fliers: ', len(bp['fliers']))
```

```python
bp = plt.boxplot(x, patch_artist=True) 

bp['boxes'][0].set(facecolor='#AFEEEE', 
                   edgecolor='blue', 
                   hatch='*')
```

```python
bp = plt.boxplot(x, patch_artist=True) 

bp['boxes'][0].set(facecolor='#AFEEEE', 
                   edgecolor='blue', 
                   hatch='*')

bp['whiskers'][0].set(color='green', 
                      linewidth=2)

bp['whiskers'][1].set(color='red')
```

```python
bp = plt.boxplot(x, patch_artist=True) 

bp['boxes'][0].set(facecolor='#AFEEEE', 
                   edgecolor='blue', 
                   hatch='*')

bp['whiskers'][0].set(color='green', 
                      linewidth=2)

bp['whiskers'][1].set(color='red')

bp['caps'][0].set(color='pink')

bp['caps'][1].set(color='purple',
                  linewidth=2)
```

```python
corn_data = pd.read_csv('Datasets/USDAProj_Corn.csv')
```

```python
corn_data.head()
```

```python
corn_data = corn_data[['World Imports', 'World Exports']]
corn_data.head()
```

```python
corn_data.describe()
```

```python
corn_data_array = np.array(corn_data)
corn_data_array
```

```python
bp = plt.boxplot(corn_data_array)

plt.show()
```

```python
bp = plt.boxplot(corn_data_array, 
                 patch_artist=True)
```

```python
print('Number of boxes: ', len(bp['boxes']))
print('Number of whiskers: ', len(bp['whiskers']))
print('Number of caps: ', len(bp['caps']))
print('Number of medians: ', len(bp['medians']))
print('Number of fliers: ', len(bp['fliers']))
```

```python
colors = ['blue', 'magenta']
```

```python
bp = plt.boxplot(corn_data_array, 
                 patch_artist=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
```

```python
bp = plt.boxplot(corn_data_array, 
                 patch_artist=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1, 2], ['World Imports', 'World Exports'])
```

```python
vp = plt.violinplot(corn_data_array)
plt.show()
```

```python
vp = plt.violinplot(corn_data_array,
                    showmedians=True)

plt.xticks([1, 2], ['World Imports', 'World Exports'])

plt.show()
```

```python
vp = plt.violinplot(corn_data_array,
                    showmedians=True,
                    vert=False)

plt.yticks([1, 2], ['World Imports', 'World Exports'])

plt.show()
```

```python
vp = plt.violinplot(corn_data_array,
                    showmedians=True,
                    vert=False)

plt.yticks([1, 2], ['World Imports', 'World Exports'])

for i in range(len(vp['bodies'])):
    
    vp['bodies'][i].set(facecolor=colors[i]) 
```

```python
vp = plt.violinplot(corn_data_array,
                    showmedians=True,
                    vert=False)

plt.yticks([1, 2], ['World Imports', 'World Exports'])

for i in range(len(vp['bodies'])):
    
    vp['bodies'][i].set(facecolor=colors[i])
                        
vp['cmaxes'].set(color='lime')

vp['cmins'].set(color='black')

vp['cbars'].set(linestyle=':')

vp['cmedians'].set(linewidth=6)
```

```python
vp = plt.violinplot(corn_data_array,
                    showmedians=True,
                    vert=False)

plt.yticks([1, 2],  ['World Imports', 'World Exports'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i]) 

plt.legend(handles = [vp['bodies'][0], vp['bodies'][1]],
           labels = ['World Imports', 'World Exports'], 
           loc = 'upper left')
```

```python
vp = plt.violinplot(corn_data_array,
                    showmedians=True,
                    vert=False)

plt.yticks([1, 2],  ['World Imports', 'World Exports'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i]) 

plt.legend(handles = [vp['bodies'][0], vp['bodies'][1]],
           labels = ['World Imports', 'World Exports'], 
           loc = 'upper left',
           bbox_to_anchor = (1,1))
```
