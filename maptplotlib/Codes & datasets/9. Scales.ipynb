```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

* Setting of a random points

```python
np.random.seed()
```

```python
y = np.random.uniform(low=0.0, high=2000, size=(2000,))
y.sort()

x = np.arange(len(y))
```

#### Linear line

  * Plotting Y vs X with the scale of Y being linear
  * Also show grid lines on our plot

```python
plt.plot(x, y)

plt.grid(True)

plt.xscale('linear')

plt.show()
```

### Logarithmic scale

  * Log value of x value is y value

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log')

plt.show()
```

* Changing the base of the log scale

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log', basey=2)

plt.show()
```

* Using different logarithm scale on each axis 

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log', basey=10)

plt.xscale('log', basex=2)

plt.show()
```
