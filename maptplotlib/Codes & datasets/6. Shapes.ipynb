```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

* We are using patch to draw shape

```python
fig, ax = plt.subplots()

ax.add_patch(
    patches.Circle(
        (0.4, 0.4),0.3,
        fill=False 
    )
)

plt.show()
```

* Using aspect parameter for the axis to fix the aspect ratio

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect='equal')

ax.add_patch(
    patches.Circle(
        (0.4, 0.4),0.3,
        fill=False 
    )
)

plt.show()
```

* giving color for the shape

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect='equal')

ax.add_patch(
    patches.Circle(
        (0.4, 0.4),0.3,
        facecolor='#00CED1',
        edgecolor='#9400D3',
    )
)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect='equal')

ax.add_patch(
    patches.Circle(
        (0.4, 0.4),0.3,
        facecolor='#00CED1',
        edgecolor='#9400D3',
        color='red'
    )
)

plt.show()
```

* Drawing multiple circles

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.4, 0.4),0.2,
        hatch='*'
    ),
    patches.Circle(
        (0.4, 0.4),0.3,
        hatch='+',
        fill=False
    ),
]:
    ax.add_patch(p)
    
plt.show()
```

* Changing the positions of circles

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.4, 0.3),0.2,
        hatch='*'
    ),
    patches.Circle(
        (0.4, 0.8),0.2,
        hatch='+',
        fill=False
    ),
]:
    ax.add_patch(p)
    
plt.show()
```

* Setting alpha for shapes

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.2, 0.3),0.2,
        alpha=None
    ),
    patches.Circle(
        (0.5, 0.8),0.2,
        alpha=1
    ),

    patches.Circle(
        (0.8, 0.3),0.2,
        alpha=0.5
    ),
    patches.Circle(
        (1.2, 0.8),0.2,
        alpha=0.3
    ),
]:
    ax.add_patch(p)
    
plt.show()
```

* Setting xlim to show all circles

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.2, 0.3),0.2,
        alpha=None
    ),
    patches.Circle(
        (0.5, 0.8),0.2,
        alpha=1
    ),

    patches.Circle(
        (0.8, 0.3),0.2,
        alpha=0.5
    ),
    patches.Circle(
        (1.2, 0.8),0.2,
        alpha=0.3
    ),
]:
    ax.add_patch(p)

ax.set_xlim(0,1.5)

plt.show()
```

* Specifying colors for circles

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.2, 0.3),0.2,
        facecolor=None
    ),
    patches.Circle(
        (0.5, 0.8),0.2,
        facecolor='none'
    ),

    patches.Circle(
        (0.8, 0.3),0.2,
        facecolor='#FF00FF'
    ),
    patches.Circle(
        (1.2, 0.8),0.2,
        facecolor='#00FA9A'
    ),
]:
    ax.add_patch(p)

ax.set_xlim(0,1.5)

plt.show()
```

* Adding edgecolor

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.2, 0.3),0.2,
        fill=False,
        edgecolor=None
    ),
    patches.Circle(
        (0.5, 0.8),0.2,
        fill=False,
        edgecolor='none'
    ),

    patches.Circle(
        (0.8, 0.3),0.2,
        fill=False,
        edgecolor='#FF00FF'
    ),
    patches.Circle(
        (1.2, 0.8),0.2,
        fill=False,
        edgecolor='#00FA9A'
    ),
]:
    ax.add_patch(p)

ax.set_xlim(0,1.5)

plt.show()
```

* Adding line width

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.2, 0.3),0.2,
        fill=False,
        linewidth=None
    ),
    patches.Circle(
        (0.5, 0.8),0.2,
        fill=False,
        linewidth=0
    ),

    patches.Circle(
        (0.8, 0.3),0.2,
        fill=False,
        linewidth=0.5
    ),
    patches.Circle(
        (1.2, 0.8),0.2,
        fill=False,
        linewidth=4    
    ),
]:
    ax.add_patch(p)

ax.set_xlim(0,1.5)

plt.show()
```

* Different types of linestyles

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Circle(
        (0.2, 0.3),0.2,
        fill=False,
        linestyle='solid'
    ),
    patches.Circle(
        (0.5, 0.8),0.2,
        fill=False,
        linestyle='dotted'
    ),

    patches.Circle(
        (0.8, 0.3),0.2,
        fill=False,
        linestyle='dashdot'
    ),
    patches.Circle(
        (1.2, 0.8),0.2,
        fill=False,
        linestyle='dashed'
    ),
]:
    ax.add_patch(p)

ax.set_xlim(0,1.5)

plt.show()
```

* Change shape style

  plotting rectangles

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Rectangle(
        (0.1, 0.1),0.1,0.6,
        hatch='.'
    ),
    patches.Rectangle(
        (0.3, 0.1),0.1,0.6,
        hatch='*',
        facecolor='#FFD700'
    ),

    patches.Rectangle(
        (0.5, 0.1),0.1,0.6,
        hatch='x',
        facecolor='#FF00FF'
    ),
    patches.Rectangle(
        (0.7, 0.1),0.1,0.6,
        hatch='+',
        Fill=False
        
    ),
]:
    ax.add_patch(p)


plt.show()
```

* Adding edgecolor

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Rectangle(
        (0.1, 0.1),0.1,0.6,
        hatch='.'
    ),
    patches.Rectangle(
        (0.3, 0.1),0.1,0.6,
        hatch='*',
        facecolor='#FFD700'
    ),

    patches.Rectangle(
        (0.5, 0.1),0.1,0.6,
        hatch='x',
        facecolor='#FF00FF'
    ),
    patches.Rectangle(
        (0.7, 0.1),0.1,0.6,
        hatch='+',
        fill=False,
        edgecolor='#00BFFF'
        
    ),
]:
    ax.add_patch(p)


plt.show()
```

* Drawing polygons using patches

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

polygon = patches.Polygon([[0.2, 0.2],
                           [0.1, 0.6], 
                           [0.4, 0.8], 
                           [0.7, 0.6], 
                           [0.6, 0.2]],
                          fill=False)

ax.add_patch(polygon)
    
plt.show()
         
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

polygon = patches.Polygon([[0.2, 0.2],
                           [0.1, 0.6], 
                           [0.4, 0.8], 
                           [0.7, 0.6], 
                           [0.6, 0.2]],
                           fill=True)

ax.add_patch(polygon)
    
plt.show()
         
```

* Specify color

* 'Open' polygon 

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

polygon = patches.Polygon([[0.2, 0.2],
                           [0.1, 0.6], 
                           [0.4, 0.8], 
                           [0.7, 0.6], 
                           [0.6, 0.2]],
                          closed=False,
                          fill=False)

ax.add_patch(polygon)
    
plt.show()
         
```

* Plot Arrows 

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

polygon = patches.Arrow(0.8, 0.1,
                        -0.5, 0.8)
                          
ax.add_patch(polygon)
    
plt.show()
         
```

* Manage arrow width

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

polygon = patches.Arrow(0.8, 0.1,
                        -0.5, 0.8,
                        width=0.6) 
ax.add_patch(polygon)
    
plt.show()
         
```

* Specify color

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

polygon = patches.Arrow(0.8, 0.1,
                        -0.5, 0.8,
                        width=0.6,
                       facecolor='#D2691E') 
ax.add_patch(polygon)
    
plt.show()
         
```
