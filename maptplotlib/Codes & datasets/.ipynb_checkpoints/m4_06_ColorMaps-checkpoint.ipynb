```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
aapl_data = pd.read_csv('AAPL.csv')
aapl_data.head()
```

```python
#aapl_data['Date'] = pd.to_datetime(aapl_data['Date'])
```

```python
a = aapl_data['Open']
b = aapl_data['Low']
c = aapl_data['High']
```

```python
x = aapl_data['Date']
```

```python
#x = np.vstack([a,b,c])
```

```python
y = np.vstack([aapl_data['Open'], 
               aapl_data['Low'],
               aapl_data['High'],])
y[0]
```

```python
labels = ['Open', 
          'Low',
          'High']
type(y)
```

```python
plt.stackplot(x, y, 
              labels=labels)

plt.legend(loc='upper left')

plt.show()
```

```python
colormap = ['magenta','pink','skyblue']
```

```python
fig, ax = plt.subplots()

ax.set_prop_cycle('color', colormap)

ax.stackplot(x, y, 
             labels=labels)

ax.legend(loc='upper left')

plt.show()
```

```python
colormap = ['magenta','skyblue']
```

```python
fig, ax = plt.subplots()

ax.set_prop_cycle('color', colormap)

ax.stackplot(x, y, 
             labels=labels)

ax.legend(loc='upper left')

plt.show()
```

```python
sp = plt.stackplot(x, y, 
              labels=labels)

plt.legend(loc='upper left')

plt.show()
```

```python
rainbow_cmap = plt.get_cmap('rainbow', len(sp))
rainbow_cmap
```

```python
fig, ax = plt.subplots()

sp = ax.stackplot(x, y, 
             labels=labels)

for i in range(len(sp)):
    sp[i].set_facecolor(rainbow_cmap(i))
```

```python
import matplotlib.colors as colors
```

```python
x = np.arange(1, 10)
y = np.log10(x)
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y, 
           cmap='rainbow')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y,
           cmap='rainbow',
           norm=colors.LogNorm())

plt.show()
```

```python
x = np.arange(1, 100)
y = np.log10(x)
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y, 
           cmap='rainbow')

ax.set_yscale('log')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y, 
           cmap='rainbow',
           norm=colors.LogNorm())

ax.set_yscale('log')

plt.show()
```
