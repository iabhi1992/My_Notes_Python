```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

* Working on interactive mode and plotting graph using x and y values

```python
plt.plot([3, 5, 7, 9],
        [6, 10,14, 18])
```

* Adding color for the line

```python
plt.plot([3, 5, 7, 9],
        [6, 10,14, 18],
        color='fuchsia')
```

* Adding x and y labels

```python
plt.plot([3, 5, 7, 9],
        [6, 10,14, 18])

plt.xlabel('x value', fontsize=18, color='blue')
plt.ylabel('y value', fontsize=18, color='lime')
```

 Plotting without specifying x coordinates

* We are giving only one list that will take as y value and x values will be the index of that values

```python
plt.plot([6, 10,14, 18])
```

```python
x = np.linspace(start=1, stop=15, num=60)
```

Generate a cos wave

* Plotting cos wave using cos fuction

```python
plt.plot(x, np.cos(x))

plt.xlabel('x value', fontsize=18, color='blue')
plt.ylabel('cos(x)', fontsize=18, color='lime')
```

* Adding tick properties

```python
plt.tick_params(axis='x',
               color='crimson',
               labelcolor='magenta',
               labelsize='large')
```

* Modifying x axis

```python
plt.tick_params(axis='y',
               left=False,
               labelleft=False)
```

```python
plt.plot(x, np.cos(x))

plt.xlabel('x value', fontsize=18, color='blue')
plt.ylabel('cos(x)', fontsize=18, color='lime')
```

* Adding legends

```python
plt.plot(x, np.cos(x), label='cosine curve')

plt.legend()
```

* Setting titles

```python
plt.plot(x, np.cos(x), label='cosine curve')

plt.legend()

plt.title('Math is Fun')
```

* Specifying x and y limits

```python
plt.plot(x, np.cos(x), label='cosine curve')

plt.legend()
plt.title('Math is Fun')

plt.xlim(4,10)
```

* Setting ylim

```python
plt.plot(x, np.cos(x), label='cosine curve')

plt.legend()
plt.title('Math is Fun')

plt.xlim(4,10)
plt.ylim(0,1)
```

* Inverting x axis

```python
plt.plot(x, np.cos(x), label='cosine curve')

plt.legend()
plt.title('Math is Fun')

plt.xlim(10,4)
plt.ylim(0,1)
```

* Inverting y axis too

```python
plt.plot(x, np.cos(x), label='cosine curve')

plt.legend()
plt.title('Math is Fun')

plt.xlim(10,4)
plt.ylim(1,0)
```
