```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
corn_data = pd.read_csv('Datasets/USDAProj_Corn.csv')
corn_data.head()
```

```python
#corn_data['Date'] = pd.to_datetime(corn_data['Date'])
```

```python
#a = stock_data['AAPL']
#b = stock_data['GOOG']
#c = stock_data['IBM']
```

```python
x = corn_data['Date']
type(x)
```

```python
y = np.vstack([corn_data['Brazil Production'], 
               corn_data['Argentina Production'],
               corn_data['European Union Production']])
type(y)
```

```python
labels = ['Brazil Production', 
          'Argentina Production',
          'EU Production']
```

```python
plt.stackplot(x, y, 
              labels=labels)

plt.ylim(0,200)
plt.legend(loc='upper left')

plt.show()
```

```python
colors = ['pink', 
          'tomato', 
          'skyblue']
```

```python
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='grey')

plt.ylim(0,200)

plt.legend(loc='upper left')

plt.show()
```

```python
aapl_data = pd.read_csv('Datasets/AAPL.csv')
aapl_data.head()
```

```python
aapl_data[['Adj Close']] = aapl_data[['Adj Close']].diff()

aapl_data.head()
```

```python
plt.figure(figsize=(10,6))

plt.stem(aapl_data['Adj Close'])

plt.title('')

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.stem(aapl_data['Adj Close'],
         markerfmt = 'g_',
         linefmt = 'r--',
         basefmt = 'b:')

plt.title('AAPL')

plt.show()
```
