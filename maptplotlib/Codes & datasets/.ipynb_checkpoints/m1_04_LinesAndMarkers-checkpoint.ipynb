```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
x = [i for i in range(1,20)]
y = x
y2 = [x**2 for x in x]
y3 = [x**3 for x in x]
```

* Plotting straight line using x and y values

```python
plt.plot(x, y)

plt.show()
```

Adding multiple plots in a pyplot

* Plotting two lines second is square of x

```python
plt.plot(x, y)
plt.plot(x,y2)

plt.show()
```

* Adding cube also

```python
plt.plot(x, y)
plt.plot(x,y2)
plt.plot(x,y3)

plt.show()
```

* Changing colour

```python
plt.plot(x, y ,color='magenta')
plt.plot(x,y2 ,color='orange')
plt.plot(x,y3 ,color='blue' )

plt.show()
```

* adding title and legend too

```python
plt.plot(x, y ,color='magenta',label='x')
plt.plot(x,y2 ,color='orange',label='x^2')
plt.plot(x,y3 ,color='blue' ,label='x^3')

plt.legend()
plt.title('Numbers')

plt.show()
```

* Using different shapes to plot

```python
plt.plot(x, y, 'r--', x, y2, 'bs', x, y3, 'g^')

plt.show()
```

* Plotting using random numbers

```python
random_array = np.random.randn(30)
```

```python
plt.plot(random_array,
         color='gold')

plt.show()
```

* Different line styles

```python
plt.plot(random_array,
         color='gold',
         linestyle='-.')

plt.show()
```

```python
plt.plot(random_array,
         color='gold',
         linestyle=':')

plt.show()
```

* Adjust the line width

```python
plt.plot(random_array,
         color='gold',
         linestyle=':',
         linewidth=5)

plt.show()
```

 Using markers to denote the points
  
 * the '*' denotes stars.
 
 * Adjusting the marker size as well

```python
plt.plot(random_array,
         color='gold', 
         marker = '*',
         markersize = 15)
plt.show()
```

* Plot using only markers

```python
plt.plot(random_array,
         color='gold', 
         marker = '*',
         linestyle='none')
plt.show()
```

* Scatter plot

```python
plt.scatter(range(0,30),
            random_array,
            color='gold')
plt.show()
```

```python
plt.scatter(range(0,30),
            random_array,
            color='gold',
            marker = '*')
plt.show()
```
