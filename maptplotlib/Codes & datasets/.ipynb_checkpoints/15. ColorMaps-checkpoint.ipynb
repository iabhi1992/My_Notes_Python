```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
corn_data = pd.read_csv('Datasets/USDAProj_Corn.csv')
corn_data.head()
```

```python
x = corn_data['Date']
```

```python
y = np.vstack([corn_data['Brazil Production'], 
               corn_data['Argentina Production'],
               corn_data['European Union Production'],])
y[0]
```

```python
labels = ['Brazil Production', 
          'Argentina Production',
          'EU Production']
type(y)
```

```python
plt.stackplot(x, y, 
              labels=labels)

plt.ylim(0,200)

plt.legend(loc='upper left')

plt.show()
```

```python
colormap = ['magenta','pink','skyblue']
```

```python
fig, ax = plt.subplots()

ax.set_prop_cycle('color', colormap)

ax.stackplot(x, y, 
             labels=labels)

plt.ylim(0,200)

ax.legend(loc='upper left')

plt.show()
```

```python
colormap = ['magenta','skyblue']
```

```python
fig, ax = plt.subplots()

ax.set_prop_cycle('color', colormap)

ax.stackplot(x, y, 
             labels=labels)

plt.ylim(0,200)

ax.legend(loc='upper left')

plt.show()
```

```python
sp = plt.stackplot(x, y, 
              labels=labels)

plt.ylim(0,200)

plt.legend(loc='upper left')

plt.show()
```

```python
rainbow_cmap = plt.get_cmap('rainbow', len(sp))
rainbow_cmap
```

```python
fig, ax = plt.subplots()

sp = ax.stackplot(x, y, 
             labels=labels)

plt.ylim(0,200)

for i in range(len(sp)):
    sp[i].set_facecolor(rainbow_cmap(i))
```

```python
x = np.arange(100)

y = x
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y, 
           cmap='rainbow')

plt.show()
```

```python
import matplotlib.colors as colors
```

```python
x = np.arange(1, 10)
y = np.log10(x)
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y, 
           cmap='rainbow')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y,
           cmap='rainbow',
           norm=colors.LogNorm())

plt.show()
```

```python
x = np.arange(1, 100)
y = np.log10(x)
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y, 
           cmap='rainbow')

ax.set_yscale('log')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, 
           c=y, 
           cmap='rainbow',
           norm=colors.LogNorm())

ax.set_yscale('log')

plt.show()
```
