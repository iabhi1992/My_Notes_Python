* Two functions truncated and tan

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0.7, 1.2, 0.5])
ax2 = fig.add_axes([0, 0, 0.8, 0.5])

ax1.plot(x, np.trunc(x),'r-o')

ax2.plot(x, np.tan(x),'b--')

plt.show()
```

* Adding labels

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0.7, 1.2, 0.5])
ax2 = fig.add_axes([0, 0, 0.8, 0.5])

ax1.plot(x, np.trunc(x),'r-o')
ax1.set_xlabel('x', fontsize=15, color='#FF1493')
ax1.set_ylabel('trunc(x)', fontsize=15, color='#FF1493')

ax2.plot(x, np.tan(x),'b--')
ax2.set_xlabel('x', fontsize=15, color='#32CD32')
ax2.set_ylabel('tan(x)', fontsize=15, color='#32CD32')

plt.show()
```

* Using different functions and plots three plots together
* Basically 2*2 contains total 4 plots

```python
fig = plt.figure(figsize=(6,6)) 

ax1 = fig.add_subplot(221)
ax1.plot([3, 6 , 8, 10], 
         [2, 5, 7, 9])
ax1.set_label('straight line')

ax2 = fig.add_subplot(222)
ax2.plot(x, np.trunc(x))

ax3 = fig.add_subplot(223)
ax3.plot(x, np.tan(x))

plt.show
```

* colspan will cover some part according to the number of span

```python
ax1 = plt.subplot2grid((2, 3), (0, 0))
ax1.plot(x, np.trunc(x))
ax1.set_ylabel('truncated')

ax2 = plt.subplot2grid((2, 3), (0, 1))
ax2.plot(x, np.tan(x))
ax2.set_ylabel('tangent')
ax2.yaxis.tick_right()

ax3 = plt.subplot2grid((2, 3), (0, 2), rowspan=2)
ax3.plot([3, 6 , 8, 10], 
         [2, 5, 7, 9])
ax3.set_xlabel('straight line')
ax3.yaxis.tick_right()

ax4 = plt.subplot2grid((2, 3), (1, 0), colspan=2)

ax4.plot(x, np.fabs(x),'-.')


ax4.set_xlabel('absolute value')
ax4.tick_params(axis='x',rotation=60)
```
