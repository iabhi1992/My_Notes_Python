```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 5.5, 8, 10],
        [3, 6, 4, 5.6, 3.5])

plt.show()
```

* Basic Annotation

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 5.5, 8, 10],
        [3, 6, 4, 5.6, 3.5])

ax.annotate('max value', 
            xy=(3, 6), 
            xytext=(3,5.0),
            arrowprops=dict(color='red'))

plt.show()
```

* Changing the text and arrow position 

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 5.5, 8, 10],
        [3, 6, 4, 5.6, 3.5])

ax.annotate('max value', 
            xy=(3, 6), 
            xytext=(5,6.0),
            arrowprops=dict(color='red'))

plt.show()
```

* Setting facecolor,edgecolor and alpha value for the arrow

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 5.5, 8, 10],
        [3, 6, 4, 5.6, 3.5])

ax.annotate('max value', 
            xy=(3, 6), 
            xytext=(5,6.0),
            arrowprops=dict(facecolor='blue', edgecolor='red', alpha=0.4))

plt.show()
```

* Adding text color

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 5.5, 8, 10],
        [3, 6, 4, 5.6, 3.5])

ax.annotate('max value', 
            xy=(3, 6), 
            xytext=(5,6.0),
            arrowprops=dict(facecolor='blue',
            edgecolor='red', alpha=0.4),
            color='magenta',style='oblique')

plt.show()
```

* Arrow annotation with point

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 5.5, 8, 10],
        [3, 6, 4, 5.6, 3.5])

ax.annotate('min value point', 
            xy=(1, 3), 
            xytext=(3,3.5),
            arrowprops=dict(color='red'))

ax.plot([1], [3], "bo")

plt.show()
```

* Using shrink parameter

  we are using shrink parameter to reduce the arrow size , So now the arrow annotation is not overlaping the point

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 5.5, 8, 10],
        [3, 6, 4, 5.6, 3.5])

ax.annotate('min value point', 
            xy=(1, 3), 
            xytext=(3,3.5),
            arrowprops=dict(color='red', shrink=0.1))

ax.plot([1], [3], "ro")

plt.show()
```

```python
import numpy as np
```

* Defining two sets of plots 

```python
np.random.seed(0)

x1 = 3 + np.random.randn(150)
y1 = 1 + np.random.randn(150)

x2 = 1 + np.random.randn(150)
y2 = 3 + np.random.randn(150)
```

* Drawing a figure using those plots

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='b')
ax.scatter(x2, y2, color='magenta')

plt.show()
```

* Defining box annotation

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='b')
ax.scatter(x2, y2, color='magenta')

bbox_props = dict(boxstyle='circle', 
                  facecolor='w', 
                  alpha=0.5)

ax.text(4, 0, 
        'Sample A', 
        ha='center', 
        va='center', 
        size=15,
        bbox=bbox_props)

ax.text(0, 4, 
        'Sample B', 
        ha='center', 
        va='center', 
        size=15,
        bbox=bbox_props)
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='b')
ax.scatter(x2, y2, color='magenta')


ax.text(4, 0, 
        'Sample A', 
        ha='center', 
        va='center', 
        size=15,
        bbox=bbox_props)

ax.text(0, 4, 
        'Sample B', 
        ha='center', 
        va='center', 
        size=15,
        bbox=bbox_props)

arrow_bbox_props = dict(boxstyle='larrow', 
                        facecolor='#EBF5FB', 
                        edgecolor='lime', 
                        linewidth=3,
                        alpha=0.7)
ax.text(2, 2, 
        'Direction', 
        ha='center', 
        va='center', 
        rotation=-30,
        size=15,
        bbox=arrow_bbox_props)
```

There are different forms of binding boxes

* Like larrow, rarrow, darrow (bidrectional arrow), circle, round (rounded rectangle), round4 (a curvier rounded rectangle), sawtooth (rectangle with sawtooth edges) etc

```python
sawtooth_bbox_props = dict(boxstyle='circle', 
                           facecolor='#ABEBC6',
                           linewidth=3,
                           edgecolor='red',
                           alpha=0.8)
```

```python
round4_bbox_props = dict(boxstyle='circle', 
                         facecolor='#EBDEF0',
                         linewidth=3,
                         edgecolor='red',
                         alpha=0.8)
```

```python
arrow_bbox_props = dict(boxstyle='darrow', 
                        facecolor='#FDEDEC',
                        alpha=0.5)
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='b')
ax.scatter(x2, y2, color='magenta')


ax.text(4, 0, 
        'Sample A', 
        ha='center', 
        va='center', 
        size=15,
        bbox=sawtooth_bbox_props)

ax.text(0, 4, 
        'Sample B', 
        ha='center', 
        va='center', 
        size=15,
        bbox=round4_bbox_props)


ax.text(2, 2, 
        'Direction', 
        ha='center', 
        va='center', 
        rotation=-30,
        size=15,
        bbox=arrow_bbox_props)
```
