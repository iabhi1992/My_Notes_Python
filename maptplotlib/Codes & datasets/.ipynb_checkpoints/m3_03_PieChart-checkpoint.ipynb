```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
gold_data=pd.read_csv('Datasets/World gold Production in 1990.csv')
gold_data
```

```python
plt.pie(gold_data['Production'], 
        labels=gold_data['Country_Name'])

plt.show()
```

```python
plt.pie(gold_data['Production'], 
        labels=gold_data['Country_Name'])

plt.axis('equal')

plt.show()
```

```python
colors = ['salmon', 'aqua', 'magenta', 'dodgerblue','lime', 'darkviolet']
```

```python
plt.pie(gold_data['Production'], 
        labels=gold_data['Country_Name'],
        colors=colors,
        autopct='%.2f')

plt.axis('equal')

plt.show()
```

```python
plt.pie(gold_data['Production'], 
        labels=gold_data['Country_Name'],
        colors=colors,
        autopct='%.2f',
        startangle=60,
        counterclock=False)

plt.axis('equal')

plt.show()
```

```python
plt.pie(gold_data['Production'], 
        labels=gold_data['Country_Name'],
        colors=colors,
        autopct='%.2f',
        startangle=180,
        counterclock=False)

plt.axis('equal')

plt.show()
```

```python
explode = (0, 0, 0, 0.1, 0, 0)
```

```python
plt.pie(gold_data['Production'], 
        labels=gold_data['Country_Name'],
        colors=colors,
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```

```python
explode = (0, 0, 0.2, 0.1, 0, 0)
```

```python
plt.pie(gold_data['Production'], 
        labels=gold_data['Country_Name'],
        colors=colors,
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```

```python
wedges, texts, autotexts = plt.pie(gold_data['Production'],
                                   labels=gold_data['Country_Name'],
                                   colors=colors,
                                   autopct='%.2f',
                                   explode=explode)

       
plt.axis('equal')

print('Wedges: ', wedges)
print('Texts: ', texts)
print('Autotexts: ', autotexts)
```

```python
wedges, texts, autotexts = plt.pie(gold_data['Production'],
                                   labels=gold_data['Country_Name'],
                                   colors=colors,
                                   autopct='%.2f',
                                   explode=explode)


plt.axis('equal')

wedges[3].set(edgecolor='red', linewidth=2)

texts[3].set(family='fantasy', size=20)

autotexts[3].set(weight='bold', size=15)
```

```python
wedges, texts, autotexts = plt.pie(gold_data['Production'],
                                   labels=gold_data['Country_Name'],
                                   colors=colors,
                                   autopct='%.2f',
                                   explode=explode)


plt.axis('equal')

wedges[2].set(edgecolor='blue', linewidth=3)

texts[2].set(family='cursive', size=20)

autotexts[2].set(weight='bold', size=13)
```
