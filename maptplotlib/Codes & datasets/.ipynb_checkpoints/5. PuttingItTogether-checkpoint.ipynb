```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

* Here we are using avacodo datasets

```python
avocado_data = pd.read_csv('Datasets/avocado.csv')
avocado_data.head()
```

* Filtering the data
* We are only using albeny region and conventional type data

```python
avocado_data = avocado_data.loc[(avocado_data['region'] == 'Albany') & (avocado_data['type'] == 'conventional')]
```

```python
avocado_data = avocado_data.sort_values(by = ['Date'])
avocado_data.head()
```

* Converting to pandas date time format

```python
avocado_data['Date'] = pd.to_datetime(avocado_data['Date'])
```

```python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_axes([0, 0, 1.5, 1.5])

ax2 = fig.add_axes([0.7, 1, 0.6, 0.4])
```

* Plotting small bags data in outside figure

```python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_axes([0, 0, 1.5, 1.5])

ax2 = fig.add_axes([0.7, 1, 0.6, 0.4])

ax1.plot(avocado_data['Date'],
         avocado_data['Small Bags'],
         color='limegreen')

ax1.set_title('Avocado Data')
```

* Plotting large bags in second figure

```python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_axes([0, 0, 1.5, 1.5])

ax2 = fig.add_axes([0.7, 1, 0.6, 0.4])

ax1.plot(avocado_data['Date'],
         avocado_data['Small Bags'],
         color='limegreen')

ax1.set_title('Avocado Data')

ax2.plot(avocado_data['Date'],
         avocado_data['Large Bags'],
         color='blue')
ax2.tick_params(axis='x',rotation=60)
```

* Plotting seperately, using subplot

```python
fig = plt.figure(figsize=(18,10))


fig.suptitle('Avocado Data',
             fontsize=25,color='#00CED1')

ax1 = fig.add_subplot(221)

ax1.set_title('AveragePrice')

ax1.plot(avocado_data['Date'],
         avocado_data['AveragePrice'], 
         color='#FF00FF')

ax1.tick_params(axis='x',rotation=60)
```

* Plotting all four data in the 2*2 grid

```python
fig = plt.figure(figsize=(18,10))


fig.suptitle('Avocado Data',
             fontsize=25,color='#00CED1')

ax1 = fig.add_subplot(221)

ax1.set_title('AveragePrice')

ax1.plot(avocado_data['Date'],
         avocado_data['AveragePrice'], 
         color='#FF00FF')

ax1.tick_params(axis='x',rotation=60)

ax2 = fig.add_subplot(222)
ax2.set_title('Total Volume')


ax2.plot(avocado_data['Date'],
         avocado_data['Total Volume'], 
         color='#00FF7F')

ax2.tick_params(axis='x',rotation=60)
```
