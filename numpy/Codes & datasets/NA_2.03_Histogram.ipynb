## Histogram

## numpy.histogram()
The NumPy <b>histogram</b> function applied to an array returns a pair of vectors: the histogram of the array and the vector of bins.


```python
import numpy as np
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
np.histogram(a,bins = [0,20,40,60,80,100]) 
```

```python
hist,bins = np.histogram(a,bins = [0,20,40,60,80,100]) 
print (hist)
print (bins)
```

## plt()
This function plot the numpy histogram data. 

```python
from matplotlib import pyplot as plt
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a, bins = [0,20,40,60,80,100]) 
plt.title("histogram") 
plt.show()
```

#### Create a normal distribution
Mean = 3
StdDev = 0.6 
Num of data points = 100

```python
mu, sigma = 3, 0.6
data = np.random.normal(mu, sigma, 100)
```

```python
data
```

```python
(n, bin_edges) = np.histogram( data, bins = 100)
```

```python
n
```

```python
bin_edges
```

```python
plt.plot(bin_edges[1 : ],n)
plt.show()
```

```python
plt.hist(data, bins = 'auto')  
plt.show()
```
