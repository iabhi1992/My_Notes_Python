```python
import numpy as np
import pandas as pd
```

```python
honeyproduction_US = pd.read_csv( './data/honeyproduction.csv')
```

https://www.kaggle.com/jessicali9530/honey-production

```python
honeyproduction_US.head()
```

```python
total_prod = np.sum(honeyproduction_US['totalprod'])
total_prod
```

```python
honeyproduction_US = honeyproduction_US.dropna(how = 'any')
honeyproduction_US.describe()
```

```python
honeyproduction_US["state"].value_counts()
```

```python
honeyproduction_US = honeyproduction_US.drop(columns = ['unnamed'])
```

```python
year_wise = honeyproduction_US.groupby('year').aggregate(np.mean)
```

```python
year_wise = year_wise.drop(columns = ['yieldpercol','priceperlb','prodvalue'])
```

```python
year_wise
```

```python
year_wise = year_wise[['totalprod']]
year_wise
```

```python
prodND = honeyproduction_US[honeyproduction_US['state'] == "ND"]
prodND
```

```python
avgprod = np.average(prodND["totalprod"])
avgprod
```

```python
year_2004 = honeyproduction_US[honeyproduction_US['year'] == 2004]
year_2004.head()
```

```python
total_prod2004 = np.sum(year_2004['totalprod'])
total_prod2004
```

```python
year_2004["prodpercentage"] = year_2004["totalprod"] / total_prod2004 * 100
```

```python
year_2004.sort_values(by = ['prodpercentage'], 
                      ascending = False)
```

```python
np.count_nonzero(year_2004['state'])
```
