```python
# Grouping means categorizing a dataframe using a certain column as a basis.
# Basis may be name, age anything
# For example - You are running an insurance company and you want to find Insurance buying pattern for different age groups.
# e.g. What is the Insurance buying pattern for a group of people of age 20-25 or 30-35?
```

```python
import pandas as pd
```

```python
cars_data = pd.read_csv(r'./datasets/cars_data.csv')
```

```python
cars_data
```

```python
groupby = cars_data.groupby('origin_city')
```

```python
groupby
```

```python
for origin_city, origin_city_df in groupby:
    print(origin_city)
    print(origin_city_df)
```

```python
groupby.get_group('Japan')
```

```python
groupby.max()
```

```python
groupby.min()
```

```python
groupby.mean()
```
