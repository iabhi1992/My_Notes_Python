```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
football_data = pd.read_csv('./Datasets/FIFA2018.csv')
```

### edgecolor and lw parameteres are to set the edgecolor of barplot and to set the thickness of the edge color. col_wrap is the parameter to set the number of plots in a single row.

```python
g=sns.FacetGrid(football_data,col='Goal Scored',
                col_wrap = 2,
                aspect = 3)
g.map(sns.barplot,'Attempts','On-Target',
      color='grey',
      edgecolor='red',
      lw=.7)
g.add_legend()
plt.show()
```

### wspace is to set horizontal space between two consecutive plots.

```python
g =  sns.FacetGrid(football_data,
                col ='Result',
                col_wrap = 3,
                aspect = 3)
g.map(sns.barplot,'Ball Possession %','Goal Scored',
      color = '#D5F5E3',
      edgecolor = 'black',lw = .5)

g.fig.subplots_adjust(wspace = 1);
```

```python
g = sns.FacetGrid(football_data,
                col ='Result',
                col_wrap = 3,
                aspect = 3)
g.map(sns.barplot,'Ball Possession %','Goal Scored',
      color = '#002090',
      edgecolor = 'black',
      lw = .5)
g.set_axis_labels('Possession','Score');
```

### By specifying the yticks the markings in the Y axis is done as 1,3,5

```python
g = sns.FacetGrid(football_data,col = 'Result',
                col_wrap = 2,
                aspect = 2)
g.map(sns.barplot,'Ball Possession %','Goal Scored',
      color = '#FA6072',
      edgecolor = 'black',
      lw = .5)
g.set_axis_labels('Possession','Goals')
g.set(yticks = [1,3,5])
plt.show()
```
