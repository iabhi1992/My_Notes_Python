```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Datasets/Pokemon.csv') 
```

The lmplot() function, a function in seaborn used to visualize a linear relationship as determined through regression. Here the plot is between the 'Speed Point' and 'Defense Point'.

```python
sns.lmplot( x = 'Speed Point',
           y ='Defense Point',
           data = pokemon_data)
plt.show()
```

#### From the above plot it can be seen that there is a relationship between Defense Point and Speed Point. The blue colour around the regression line shows the confidence interval of the data. Where the the colour strip is converging there the confidence interval is smaller and viceversa.

```python
sns.lmplot( x = 'Attack Point',
           y ='Defense Point',
           data = pokemon_data)
plt.show()
```

#### In this plot the with the attribute 'size' and 'aspect' the size and the aspect ratio of the the plot is set respectively.

```python
sns.lmplot( x = 'Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           size = 3,
           aspect = 2)
plt.show()
```

#### For the central tendency approach, we need a statistical function that can estimate the mean value of the given set of values. This can be governed by the np.mean() function.

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           x_estimator = np.mean)
plt.show()
```

#### In the following plot the 'lmplot()' function  is used for different values for the  column     'Type 1'.

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
          col = 'Type 1')
plt.show()
```

The following operation is similar like the above one but one additional attribute 'hue' is added, and it is set to 'Type 2'. Hue can be considered as dimensions of x and y.

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           col = 'Type 1',
           hue = 'Type 2')
plt.show()
```

These are plots between 'Attack Point' and 'Defense Point for different 'Type 1' values and the hue,'Type 2' is denoted with the colours which are assigned separately for every 'Type 2' values.

#### Only by the function sns.regplot() the regression plot along with the scatter plot can be visualised.

```python
sns.regplot(x ='Attack Point',
            y ='Defense Point',
            data = pokemon_data)
plt.show()
```

The x and y axis can also be specified as given below.

```python
sns.regplot(x= pokemon_data['Attack Point'],
            y= pokemon_data['Defense Point'])
plt.show()
```

By plt.subplots() function of matplotlib the fig size can be set. Here the figure size is 7 along the x direction and 6 along the y direction.

```python
plt.subplots(figsize=(7,6))
sns.regplot(x = 'Attack Point',
            y ='Defense Point',
            data = pokemon_data)
plt.show()
```

By specifying the 'Kind' attribute to 'reg', a joint plot with regression line can be visualised.

```python
sns.jointplot(x="Attack Point",
              y="Defense Point",
              data=pokemon_data,
             xlim={0,100},
              ylim={0,70},
              kind='reg')
```

```python
sns.pairplot(pokemon_data,
            size=5,aspect=0.9,
            x_vars=['Attack Point','SpeedPoint ','Defense Point'],
            y_vars=['Total Point'],
            kind='reg')
plt.show()
```
