```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

The following dataset has different records of all the soccer teams such as goal scored,Ball possession %, Attempts,On-Target,Off-Target etc. Different types of statistiacl anslysis are being done on the dataset.

```python
soccer_data = pd.read_csv('./Datasets/FIFA2018.csv')
soccer_data.head()
```

```python
soccer_data.describe()
```

FacetGrid is a subplot grid for plotting conditional relationships.

* Visualising distributions and relations between variables, within subsets of the concerned dataset.
* As seen before, there are three dimensions along which we observe the data - row, column and hue.
* We first initialise a FacetGrid object with a dataframe
* Specify the row, column and hue parameters
* Every unique value of the categorical variable is then used to create a facet along the specified axis.
* A plotting function is applied to each subset by using a map function

### Here the column is specified as 'Result' and the histogram is plotted with the data 'Goal Scored'.

```python
g = sns.FacetGrid(soccer_data,
                  col='Result')
g.map(plt.hist,'Goal Scored')
plt.show()
```

### Now multiple values are used to make a relational plot.
Here scatter plot is used for bivariant data.

```python
g = sns.FacetGrid(soccer_data,
                  col="Result")
g.map(plt.scatter,'Goal Scored','Attempts')
plt.show()
```

### Here the legend is added with the function '.add_legend()'

```python
g=sns.FacetGrid(soccer_data,
                  col="Result",
                  hue='On-Target')
g.map(plt.scatter,'Goal Scored','Attempts')
g.add_legend()
plt.show()
```

#### The 'margin_titles'  parameter is made true to add the title for each row

```python
g=sns.FacetGrid(soccer_data,row='Result',
                col='Pass Accuracy %',
                hue ='On-Target',
                margin_titles = True)
g.map(sns.regplot,'Goal Scored','Attempts',fit_reg = False)
g.add_legend()
plt.show()
```

'gridspec_kws' is added to set the width ratios for the columns. 

```python
g=sns.FacetGrid(soccer_data,row='Result',
                  col='Goal Scored',
                  hue='Yellow Card',
                  margin_titles=True,
                gridspec_kws={'width_ratios':[4,4,4,4,4,4,4]})
g.map(sns.regplot,'Pass Accuracy %','Attempts',fit_reg=False)
g.add_legend()
plt.show()
```

#### By creating a dicyionary, colours can be assigned to  different hue.<br/>
Here the 'pal' dictionary is defined and that is passed through the 'palette' attribute.

```python
pal={'Win':'red','Lose':'green','Draw':'gray'}
g=sns.FacetGrid(soccer_data,row='Goal Scored',
                  col='Corners',
                  hue='Result',
                  margin_titles=True, 
                  palette=pal)
g.map(sns.regplot,'Pass Accuracy %','Ball Possession %',fit_reg=False)
g.add_legend()
plt.show()                 
```

#### By creating a dicyionary, shapes can be assigned to  different hue.<br/>
Here the 'i' dictionary is defined and that is passed through the 'hue_kws' attribute.

```python
pal={'Win':'red','Lose':'green','Draw':'gray'}
i={'marker':["v","^","o"]}
g=sns.FacetGrid(soccer_data,
                  row='Goal Scored',
                  col='Corners',
                  hue='Result',
                  margin_titles=True,
                  palette=pal,
                  hue_kws=i)
g.map(sns.regplot,'Pass Accuracy %','Ball Possession %',fit_reg=False)
g.add_legend()
plt.show()     
```

#### Here the Barplot is being plotted for between Attempts and Goal Scored for results column.

```python
g=sns.FacetGrid(soccer_data,col="Result",aspect=3)
g.map(sns.barplot,'Attempts','Goal Scored')
g.add_legend()
plt.show()
```

#### Among the plots the highest amount of attempt and the goal scored can be seen where the result is 'win'.

```python
g=sns.FacetGrid(soccer_data,col="Result")
g.map(sns.barplot,'Attempts','Blocked')
g.add_legend()
plt.show()
```

```python
g=sns.FacetGrid(soccer_data,col="Result")
g.map(sns.barplot,'Attempts','On-Target')
g.add_legend()
plt.show()
```
