#### Finding correalations in Real Data

```python
crashes = sns.load_dataset('car_crashes')
sns.regplot(x = 'alcohol', y = 'total',data = crashes)
sns.jointplot(x = 'alcohol', y = 'total',data = crashes,kind = 'reg');
```
