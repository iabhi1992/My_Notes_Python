```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

### Seaborn is a plotting library

```python
hist = sns.distplot(tips['total_bill'])
hist.set_title('Total Bill Histogram with Density plot')
plt.show()
```

#### Here we can see the sns.distplot() function plots a histogram along with the kernal density esrtimation plot. KDE smooths the histogram if we only wish to get the histogram only that we will add aparameter called KDE = false to the function.

```python
hist = sns.distplot(tips['total_bill'], kde = False)
hist.set_title('Total Bill Histogram')
hist.set_xlabel('Total Bill')
hist.set_ylabel('Frequency')
plt.show()
```

#### If the requirement is only the kde diagram then the histogram can be set to false. In the histogram the graph is with the raw frequency count but in kde the plot is with respectt to probability o to one.

```python
hist = sns.distplot(tips['total_bill'], hist = False)
hist.set_title('Kernel Density plot')
hist.set_xlabel('Total Bill')
hist.set_ylabel('Unit Probability')
plt.show()
```

```python
hist = sns.distplot(tips['total_bill'], rug = True)
hist.set_title('Total Bill Histogram with Density and Rug plot')
hist.set_xlabel('Total Bill')
plt.show()
```

#### Bar plot or the count plot are being created the countplot() function.

```python
count = sns.countplot('day',data = tips)
count.set_title('Count of days')
count.set_xlabel('Day of the Week')
count.set_ylabel('Frequency')
plt.show()
```

### plots for Bivariant data or data having two variable.

```python
scatter = sns.regplot(x = 'total_bill', y = 'tip', data = tips)
scatter.set_title('Scatterplot of Total Bill and Tip')
scatter.set_xlabel('Total Bill')
scatter.set_ylabel('Tip')
plt.show()
```

#### The lm plot is a another function to plot scatter diagram. The lm method plots in terms of figure object where as the regplot represents the data in terms of axes.

```python
fig = sns.lmplot( x = 'total_bill', y = 'tip', data = tips)
scatter.set_title('Scatter plot of the Total Bill and Tip')
plt.show()
```

```python
scatter = sns.jointplot( x = 'total_bill', y = 'tip', data = tips)
scatter.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
scatter.fig.suptitle('Joint plot of Total Bill and Tip')
plt.show()
```

```python
hbin = sns.jointplot( x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
hbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
hbin.fig.suptitle('Hexbin Joint plot of the Total Bill and Tip')
plt.show()
```

```python
kde = sns.kdeplot(data = tips['total_bill'], data2 = tips['tip'], shade = True)
kde.set_title ('Kernel Density of Total Bill and Tip')
kde.set_xlabel('Total Bill')
kde.set_ylabel('Tip')
plt.show()
```

```python
kde_jointplot = sns.jointplot( x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
bar = sns.barplot(x = 'time', y = 'total_bill',data = tips)
bar.set_title('Barplot of average total bill for time of day')
bar.set_xlabel('Time of day')
bar.set_ylabel('Average total bill')
plt.show()
```

```python
box = sns.boxplot(x = 'time', y = 'total_bill',data = tips)
bar.set_title('Box plot of average total bill by time of day')
box.set_xlabel('Time of day')
box.set_ylabel('Total bill')
plt.show()
```

#### To know how skwed the data is violin plot can be used as the boxplot dont include the outliers.

```python
violin = sns.violinplot(x = 'time',y = 'total_bill', data = tips)
violin.set_title('violin plot of total bill by time of day')
violin.set_xlabel('Time of day')
violin.set_ylabel('Total Bill')
```

```python
tips = tips[['total_bill','tip','size']]
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot,rug = True)
plt.show()
```
