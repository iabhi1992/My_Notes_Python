```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Datasets/Pokemon.csv')
```

sns.distplot() function shows the histogram and Kernel Density Estimation curve for a univariant data without including any attribute. From the following operation on the 'Attack Point' column, we can see that the most of the pokemon's Attack points lie in the range of 50-80 and from the kde curve it can be seen that the highest no of Pokemons are having Attack point around 60.

```python
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

The colour of the histogram and the kde can be changed just by including the attribute 'color' and the function '.set_title()', the Title of the histogram can be set.

```python
hist = sns.distplot(pokemon_data['Attack Point'] , color = 'black')
hist.set_title('Attack capability')
plt.show()
```

If the kde is not desired then the kde is set to the False. To add label in the x and y axis the functions '.set_xlabel' and '.set_ylabel' are used respectively.

```python
hist = sns.distplot(pokemon_data['Attack Point'], kde = False)
hist.set_title('Attack capability')
hist.set_xlabel('Attack Point')
hist.set_ylabel('Frequency')
plt.show()
```

Similarly like above if the histogram is not desired then 'hist' attribute is set to be false.

```python
hist = sns.distplot(pokemon_data['Attack Point'], hist = False)
hist.set_title('Kernel Density plot')
hist.set_xlabel('Attack')
hist.set_ylabel('Unit Probability')
plt.show()
```

Rug plot can be visulased from the data just by adding the 'rug' attribute and setting it to True.

```python
hist = sns.distplot(pokemon_data['Attack Point'], rug = True)
hist.set_title('Attack capability with Density and Rug plot')
hist.set_xlabel('Attack')
plt.show()
```

sns.rugplot() function directly gives the rug plot and the height can be changed from the 'height' attribute and with the 'axis' attribute the plotting axis is set.

```python
sns.rugplot(pokemon_data['Attack Point'],
            height = 0.5,
            axis = 'x')
plt.show()
```

```python
sns.distplot(pokemon_data['Attack Point'],
             bins = 40,
             kde = False,
             rug = True)
plt.show()
```

Bins divide the entire range of values into a series of intervals. The higher the number of bins the larger variation in the histogram can be seen. And the kde is being set to false.

```python
sns.distplot(pokemon_data['Attack Point'],
             bins = 75,
             kde = False,
             rug = True)
plt.show()
```

Here the histogram is not desired so it is set to false.

```python
sns.distplot(pokemon_data['Attack Point'],
             bins = 40,
             hist = False,
             rug = True)
plt.show()
```

The 'Shade' attribute is set to 'True'
in order to create a shaded region.

```python
sns.set(color_codes = True)
sns.kdeplot(pokemon_data['Attack Point'],
            shade = True,
            color = 'g')
plt.show()
```

The bandwidth is set by the 'bw' attribute and writing the desired values. The label can be added by 'label' attribute and equating it to the bandwidth value.

```python
hist = plt.subplots(figsize=(20,5))
sns.kdeplot(pokemon_data['Attack Point'])
sns.kdeplot(pokemon_data['Attack Point'],bw = 0.7,label ='bw:0.7')
sns.kdeplot(pokemon_data['Attack Point'],bw = 6,label ='bw:6')
sns.kdeplot(pokemon_data['Attack Point'],bw = 30,label ='bw:30')
sns.kdeplot(pokemon_data['Attack Point'],bw = 110,label ='bw:110')
plt.show()
```

We can make a few observations from the obtained plot.
* Smaller the bandwidth parameter, more accurate is the curve. But too small a bandwidth unnecessarily crowds the datapoints
* When explicitly not specified, the tool automatically choses the bandwidth it thinks best describes the distribution, which, more often than not, turns out pretty close.
* While estimating to draw the curve, the path traces points that are beyond the range of the dataset.
* The bw parameter controls how tightly the estimation is fit to the data.


### sns.jointplot() fubnction is used to plot bivariant data i.e data with two variables.
The two variables of the scatter plot is 'Attack Point' and 'Defense Point' as well as the histograms are plotted for both of the variables.

```python
sns.jointplot(x ='Attack Point',
              y ='Defense Point',
              data = pokemon_data)
plt.show()
```

The x limit and y limit can be changed by the attributes 'xlim' and 'ylim'.

```python
sns.jointplot(x = 'Attack Point',
              y ='Defense Point',
              data = pokemon_data,
              xlim = {0,450},
              ylim = {0,200})
plt.show()
```

In the folowing plot the scatter plot has been replaced by the attribute 'kind' and equatting it to 'hex'. This 'hex' keyword produces Hexbin plot or in other words, the bivariate analogue of a histogram. 

```python
with sns.axes_style('white'):
    sns.jointplot(x ='Attack Point',
                  y ='Defense Point',
                  data = pokemon_data,
                  kind ='hex',
                  color ='b')
```

From the above Hexbin plot it can be observed that the contrast of the hexbin is maximum corresponding to the high values of both Attack and Defense Point.

#### In the folllowing plot is kde plot for bivariant data.

```python
with sns.axes_style('white'):
    sns.jointplot(x ='Attack Point',
                  y ='Defense Point',
                  data = pokemon_data,
                  kind ='kde')
```

It can be observed that that the colour contrast of the kde is the maximum corresponding to the high values of both Attack and Defense Point which can be seen from the univariant kde plots.

#### The colour has been set to 'summer' by the 'color' attribute.

```python
sns.kdeplot(pokemon_data['Attack Point'],
            pokemon_data['Defense Point'],
            color = 'summer')
plt.show()
```

#### By the 'clip' attribute the x and y limit can be set.

```python
sns.kdeplot(pokemon_data['Attack Point'],
            pokemon_data['Defense Point'],
            clip = (0,100))
plt.show()
```

By the matplotlib functions 'plt.xlim' and 'plt.ylim' the limit for x and y can be set.

```python
sns.kdeplot(pokemon_data['Attack Point'],
            pokemon_data['Defense Point'])
plt.xlim(0,750)
plt.ylim(0,500)
```

#### Here the rug plot is plotted for both of the variables. The colour is set and the height is changed with the 'color' and the 'height' attribute respectively.

```python
sns.kdeplot(pokemon_data['Attack Point'],
            pokemon_data['Defense Point'])
sns.rugplot(pokemon_data['Attack Point'],
            color = 'r',
            height = .05)
sns.rugplot(pokemon_data['Defense Point'],
            color = 'b',
            vertical = True,
            height = .05)
plt.show()
```

#### dataframe.corr() function is used to compute pairwise correlation of the columns. </br>
#### sns.heatmap() function is to get the heatmap after finding out the correalation between the variables, by setting the 'square' and 'annot' attribute to true, the heat map can be visualised in square and the correlation value can be seen respectively. The 'fmt' attribute is to get the values after decimal point. If it is set to '.2f' then number of digits after the decimal will be two.</br>
#### The 'cmap' attribute is used to change the colour map of the heatmap.

```python
cormat = pokemon_data.corr()
plt.subplots(figsize = (7,7))
sns.heatmap(cormat, vmax = .8,
            square = True,
            annot = True, 
            fmt ='.2f', 
            cmap = 'Blues')
plt.show()
```

#### From the above heatmap it can be seen that the Sp. Atk Point and the Sp. Def Point are highly correlated with the correaltion value 0.82.
