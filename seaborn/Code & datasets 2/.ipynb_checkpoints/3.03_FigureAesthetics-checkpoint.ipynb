```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Datasets/Pokemon.csv') 
print(pokemon_data.head())
```

```python
sns.axes_style()
```

The above command dumps out the details of the currently enabled axes style.

The parameters that are part of the style definition can be overridden by passing a dictionary of parameters as below:

```python
plt.subplots(figsize=(10,5))
sns.distplot(pokemon_data['Total Point'])
plt.show()
```

Above is the plot with the basic style setting. Let us now try varying some parameters to see the effect on our plot.

```python
sns.set_style('ticks',{'xtick.major.size':8,
                      'xtick.minor.size':4.0,
                      'ytick.color':'r',
                      'xtick.color':'r',
                      'axes.facecolor':'c',
                      'ytick.direction':'in'})

f,ax=plt.subplots(figsize=(10,5))
sns.distplot(pokemon_data['Total Point'])
plt.show()
```

We see that the complete control is in our hands. We can make the graph look however we think is suitable for our application.

### Scaling plot elements

Seaborn provides the flexibility to control the scale of plot elements, which means we can use the same code to simply create appropriatey sized plots.

```python
#Resetting all parameters to defaults
sns.set()
```

There are four preset contexts available on Seaborn -  paper mode,notebook mode,talks and posters mode.

```python
sns.set_context('paper')

f,ax=plt.subplots(figsize=(10,5))
sns.distplot(pokemon_data['Total Point'])
plt.show()
```

```python
sns.set_context('talk')

f,ax=plt.subplots(figsize=(10,5))
sns.distplot(pokemon_data['Total Point'])
plt.show()
```

```python
sns.set_context('poster')

f,ax=plt.subplots(figsize=(10,5))
sns.distplot(pokemon_data['Total Point'])
plt.show()
```

```python
sns.set_context('notebook')

plt.subplots(figsize=(10,5))
sns.distplot(pokemon_data['Total Point'])
plt.show()
```
