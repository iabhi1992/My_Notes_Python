```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

How to create univariable statistical plot, here the plot is histogram

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'], bins = 10)
axes1.set_title('Histogram of the total bill')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Frequency')
fig.show()
```

### Bivariant statistical plot - 1. Scatter plot

### Scatter plots are used when two continious variables are being plotted against one another

first thing is to initialize the figure than the axesw and the axis

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])# scatter plots are visualisation between two continious variable
axes1.set_title('Scatterplot of Total bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show()
```

### Box plots are being used when dicrete variables are needed to be plot against one variable.

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],tips[tips['sex'] == 'Male']['tip']], labels = ['Female','Male'])
axes1.set_title('Boxplot of Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
boxplot.show()
```

### How to plot multivariant data using matplotlib

```python
def recode_sex(sex):
    if sex == 'Female':
             return 0
    else:
             return 1
tips['sex_color'] = tips['sex'].apply(recode_sex)
print(tips.head())
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(x = tips['total_bill'], y = tips['tip'], s = tips['size']*10, c = tips['sex_color'],alpha = 0.5)
axes1.set_title('Total Bill vs Tip color by Sex and sizes by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show()
```
