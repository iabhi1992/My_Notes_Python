```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

### Pairplot grid plots pairwise relationship between datasets.

```python
g=sns.PairGrid(tips,
              vars=['total_bill','tip','size'])
g.map(plt.scatter);
```

The diagonal plots are univariant and the non-diagonal plots are showing bivariant relationship.

### Here is  .map(plt.scatter) is plotting the non-diagonals plot in scatter and  .map_diag(sns.kdeplot) is plotting the diagonal plots are in kde.

```python
g=sns.PairGrid(tips,
              vars=['total_bill','tip','size'])
g.map(plt.scatter)
g.map_diag(sns.kdeplot)
```

```python
g=sns.PairGrid(tips,
              vars=['total_bill','tip','size'],
              hue='sex')
g.map(plt.scatter)
g.map_diag(sns.kdeplot)
plt.show()
```

### Here all the upper diagonal, lower diagonal and diagonal plots are set as scatter,regression and kde plot respectively.

```python
g=sns.PairGrid(tips,
              vars=['total_bill','tip','size'])
g.map_diag(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_lower(sns.regplot)
plt.show()
```

```python
g=sns.PairGrid(tips,
              x_vars=['tip','total_bill'],
              y_vars=['size','tip'],
              hue='sex')
g.map(plt.scatter)
g.add_legend()
plt.show()
```
