Even though seaborn is built upon matplotlib but there are still some matplotlib command we need to run to render our plot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(anscombe.x,anscombe.y)
plt.show()
```

matplotlib has connected each point row by row,instead of connecting the points only the plotting of points are desireable.

In order to do that the letter 'o' is being introduced in the code and this is the way to introduce points in matplotlib.

```python
plt.plot(anscombe.x,anscombe.y,'o')
plt.show()
```

### fig[axes[axis]]

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
axes1.plot(dataset_1.x, dataset_1.y,'o')
axes2.plot(dataset_2.x, dataset_2.y,'o')
axes3.plot(dataset_3.x, dataset_3.y,'o')
axes4.plot(dataset_4.x, dataset_4.y,'o')
axes1.set_title('dataset1')
axes2.set_title('dataset2')
axes3.set_title('dataset3')
axes4.set_title('dataset4')
fig.suptitle('Anscombe Data')
fig.tight_layout() # Tweak spacing to prevent clipping of ylabel
plt.show()
```

All of the descriptive statistics looks the same but when the datasets are being plotted then the differenece can be seen properly. Sometimes just seeing the descriptive statistics of the data is not enough so data visualisation is important to understand the data properly.
