```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Such plots, where one of the variables of interest is categorical, are best approached by methods specific to categorical plots. Seaborn provides many functions to analyse such data:
* swarmplot() and stripplot() for plots that show each observation at each level of the categorical variable
* boxplot() and violinplot() for abstract representations of each distribution of observations
* barplot() and pointplot() for those that apply statistical estimation to show a measure of central tendency and confidence interval.
* factorplot(), that takes the help of FacetGrid and allows the user to plot across a grid of figure panels, unlike the rest that plot onto a specific matplotlib axis.

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

* sns.set() is for setting style for better visibility in seaborn.
* np.random.seed() is to set up for statistical analysis.

```python
sns.set(style='whitegrid',
        color_codes = True)
np.random.seed(sum(map(ord,'categorical')))
```

### Strip plot

```python
sns.stripplot(x='day',
              y='tip',
              data = tips,
              size=6)
plt.show()
```

We see that stripplot is basically a scatterplot to the case where one variable is categorical. It is clear that points overlap though.

The size and the figure sizes are cahanged to see any significant cahnges in the visualisation.

```python
plt.subplots(figsize=(15,5))
sns.stripplot(x='day',
              y='tip',
              data=tips,
              size=6)
plt.show()
```

#### The 'jitter' attribute is set to true to make the data points in a dispersed. It will avoid the overlapping of the data points.

```python
plt.subplots(figsize=(15,5))
sns.stripplot(x='day',
              y='tip',
              data=tips,
              size=6,
              jitter=True)
plt.show()
```

### Swarm plot<br/>
The swarm plot output is spread out such that overlap is minimum. It is a true measure of the number of samples being considered.

```python
plt.subplots(figsize=(15,5))
sns.swarmplot(x='day',
              y='tip',
              data=tips)
plt.show()
```

From the swarm plot following points can be observed:

* Colours in the above plot are random.

* In a categorical distribution, the categorical values are arranged in ascending order along the axis. If the data is in string format, the values appear in the order specified in the dataframe.

The 'hue' attribute is set to 'size'.

```python
plt.subplots(figsize = (15,5))
sns.swarmplot(x ='day',
              y ='tip',
              data = tips,
              hue ='size')
plt.show()
```

From the above plot the number of size of the groups can be seen for different tips given in different days. Different colours has been assigned to different group size.

#### In the following the Swarm has been plotted horizontally.

```python
plt.subplots(figsize =(15,5))
sns.swarmplot(x = 'tip',
              y ='day',
              data = tips,
              hue ='size')
plt.show()
```

### Box plot<br/>
A simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles.<br/>
The lines which are being extended in both of the sides are called whiskers. The whiskers extend to points that lie within 1.5 times the interquartile range of the lower and upper quartile. Observations falling outside this margin are displayed independently. Which are called outliers.

```python
plt.subplots(figsize=(15,5))
sns.boxplot(x='day',
            y='tip'
            ,data=tips)
plt.show()
```

### Violin Plot
* The name comes from the shape of the graph plotted. We see that   this is similar to the box plot, except that the shaded region is enclosed between curves.
* The representation provides a better description of the values     of the distribution.
* Linking this to what we observed while analysing univariate distributions, we see that these curve boundaries correspond to the kernel density estimate
* The quartile and whisker values are part of the violin itself.

```python
plt.subplots(figsize=(15,5))
sns.violinplot(x='day',
               y='tip',
               data=tips)
plt.show()
```

* Setting the 'scale' as count scales the width of the violins is being equated to the number of observations in that bin.
* The 'bw' is set to .1 to get more precise distribution of points.

```python
plt.subplots(figsize=(15,5))
sns.violinplot(x ='day',
               y ='tip',
               data = tips, bw=.1,
               scale='count')
plt.show()
```

The visualisation inside the plot is being changed by setting the <b>'inner'</b> attribute to <b> 'stick'</b> and the colour has been changed by equating the <b>'palette'</b> attribute to <b>'Set3'</b>.

```python
f,ax=plt.subplots(figsize=(15,5))
sns.violinplot(x ='day',
               y ='tip',
               data = tips,
               inner = 'stick',
               palette ='Set3')
plt.show()
```

The swarm plot is being showen inside the violin plot.

```python
plt.subplots(figsize = (15,5))
sns.violinplot(x ='day',
               y = 'tip',
               data = tips,
               inner = None)
sns.swarmplot(x ='day',
              y ='tip',
              data = tips,
              color ='w')
plt.show()
```

### Statistical estimation within categories
**Plots to show the central tendency of the values in the distribution**
Here the bars are being plotted between the tip and the data by specifying the 'hue' as 'size', the realtion between the tip and the days can be seen for different group sizes.

```python
plt.subplots(figsize = (15,5))
sns.barplot(x ='day',
            y ='tip',
            data = tips,
            hue='size')
plt.show()
```

#### To show the number of variables in each category - countplot() is used.

```python
plt.subplots(figsize=(15,5))
sns.countplot(x='tip',
              data=tips,
              palette='Blues_d')
plt.xticks(rotation=90)
```

Here we can clearly see that the tip of $2 is given for maximum amount of time.

```python
f,ax=plt.subplots(figsize=(15,5))
sns.pointplot(x ='day',
              y ='tip',
              data = tips)
plt.show()
```

We see that the plot is a set of point estimates and confidence intervals. The line is connected through their mean and the length of the vertical lines denotes the confindence interval, the longer the line the higher the confidence interval and viceversa.

```python
f,ax=plt.subplots(figsize=(15,5))
sns.pointplot(x='day',
              y='tip',
              data=tips,
              hue='size')
plt.show()
```

* Although the above graph is too complicated to understand, we notice that pointplot connects points from the same hue category. This feature can be used to study how the main relationship is changing as a function of a second variable.

#### Here the box is plotted horizontally by specifying the 'orient' attribute as 'h'.

```python
plt.subplots(figsize=(15,5))
sns.boxplot(data = tips,
            orient ='h')
```

### Factorplot <br/>
A factor plot is simply the same plot generated for different response and factor variables and arranged on a single page. The underlying plot generated can be any univariate or bivariate plot.
### sns.factorplot() is the function for seaborn

```python
sns.factorplot(x='day',
               y='tip',
               data=tips,
               hue='size',
               size=5,
               aspect=2,
              kind='bar')
plt.show()
```

In the following factor plot the day vs tip box plots are plotted for different sizes of group.

```python
sns.factorplot(x='day',
               y='tip',
               data=tips,
               col='size',
               size=5,aspect=2,
               kind='box')
plt.show()
```

```python
sns.factorplot(x='day',
               y='tip',
               data=tips,
               col='size',
               size=5,aspect=2,
               kind='swarm')
plt.show()
```
