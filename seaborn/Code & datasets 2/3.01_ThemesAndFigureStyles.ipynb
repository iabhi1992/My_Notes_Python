```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Datasets/Pokemon.csv')  
print(pokemon_data.head())
```

```python
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

#### sns.set() is setting the seaborn to its default.

```python
sns.set()
```

```python
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

As soon as we set the plot to its default and plotted the graph again, we notice the change in pattern. Shows that our notebook was not in the default mode earlier

Seaborn has five preset themes - darkgrid, whitegrid, dark, white and ticks. Let us see each of these one by one.

```python
sns.set_style('whitegrid') #Grid with white background
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

```python
sns.set_style('darkgrid') # Grid with dark background
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

```python
sns.set_style('dark') # Only Dark background
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

```python
sns.set_style('white') # Only white background
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

```python
sns.set_style('ticks') # Ticks in both the axis with white background
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

```python
#For temporarily setting the style - Use the 'with' keyword
with sns.axes_style('darkgrid'):
    plt.subplots(figsize = (10,5))
    sns.distplot(pokemon_data['Attack Point'])
    plt.show()
```

The sns.despine() will set the x axis nad y axis in bottom and left side of the plot by default.

```python
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
sns.despine()
plt.show()
```

By specifying the left,right,top and bottom arguments inside the sns.despine() function we can change the position of the x and y axis.

```python
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
sns.despine(left = True,
            right = False,
            bottom = True,
            top = False)
plt.show()
```

The offset argument in the sns.despine() function sets a  space between the axis and the spine by specifying it the spines move inward from the axis. 

The trim parameter will limit the range of the surviving spines.

```python
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
sns.despine(offset =15,
            trim = True)
plt.show()
```
