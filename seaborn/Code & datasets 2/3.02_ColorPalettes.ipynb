```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Seaborn provides the option to use colours for visualising data, through a variety of palettes.

```python
pokemon_data = pd.read_csv('Datasets/Pokemon.csv')
```

### sns.set_palette() function is to set the color of the plot. Here the palette is being set to 'ocean_r'

```python
sns.set_palette('ocean_r')
plt.subplots(figsize = (10,5))
sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

### Here the style has been set by sns.set_style() function and temporary palette has been assigned to it.

```python
sns.set_style('white')
#Using a temporary palette,
sns.set_palette('Reds')
plt.subplots(figsize = (10,5))

sns.distplot(pokemon_data['Attack Point'])
plt.show()
```

Depending on the kind of data you want to plot, you can choose from different categories of palettes:
* Qualitative palettes for categorical data
* Sequential or Diverging plots for large range of data 

## Qualitative Color Palettes

These palettes are mostly useful for plots that have categorical data, or data that varies in discrete chunks without an inherent ordering. It comes with 6 variations of default themes.

```python
#Taking the Seaborn setup to its default settings
sns.set()
```

Let us use the Seaborn function palplot() to plot the values in the color palette as a horizontal array

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

The six variations are deep, muted, pastel, bright, dark and colorblind.

### Here the color palette variations are helping to show the kde plot of different bandwidths.

```python
#setting the size for the plot
plt.subplots(figsize = (15,5))

sns.set_context('poster')

sns.kdeplot(pokemon_data['Attack Point'],bw=.04,label='bw:0.04')
sns.kdeplot(pokemon_data['Attack Point'],bw=1,label='bw:1')
sns.kdeplot(pokemon_data['Attack Point'],bw=2,label='bw:2')
sns.kdeplot(pokemon_data['Attack Point'],bw=5,label='bw:5')
sns.kdeplot(pokemon_data['Attack Point'],bw=7,label='bw:7')
sns.kdeplot(pokemon_data['Attack Point'],bw=9,label='bw:9')
plt.show()
```

#### If there is plot where there are more than 6 lines are needed then  in the function sns.color_palette() the number of colors needed can be specified in the hls space.

```python
sns.palplot(sns.color_palette('hls',12))
plt.show()
```

#### Here, we are setting the lightness and saturation content of the palette by the arguments l and s.

```python
sns.palplot(sns.hls_palette(12,l = 0.2, s = 0.5))
plt.show()
```

#### It Helps for choosing a qualitative panel.

```python
sns.choose_colorbrewer_palette('qualatative')
```

#### We see that the pairplot has pairs of colours that form the color panel. Observe that the exact shades shown above are the ones used in the graph below.

```python
sns.pairplot(pokemon_data,
            vars=['Attack Point','Defense Point','Speed Point'],
            hue='Type 1',
            palette='Dark2')
plt.show()
```

### By listing out the colors can be specified explicitly.

```python
sns.palplot(sns.color_palette(['#7b59b6','#6498db','#e54c3c',
                               '#35495e','#2ecc71']))
```

### Here we can see the different colours for the different bandwidths.

```python
plt.subplots(figsize=(15,5))

sns.xkcd_rgb['denim blue']
sns.kdeplot(pokemon_data['Attack Point'],bw=.04,label='bw:0.04')
sns.xkcd_rgb['medium green']
sns.kdeplot(pokemon_data['Attack Point'],bw=1,label='bw:1')
sns.xkcd_rgb['pale red']
sns.kdeplot(pokemon_data['Attack Point'],bw=2,label='bw:2')
plt.show()
```

## Sequential Color Palettes
These palettes are most useful for plots where the data has a wide range, but the relatively low values are uninteresting, and the relatively high values are interesting.

* These are very commonly used as colormap in functions like kdeplot() or such matplotlib functions.
* Best to use palettes that have a relatively subtle shift in hue, accompanied by a large shift in brightness and saturation. Important data is highlighted well.
* You can choose the Color Brewer library for this too.

```python
#Using the Blues palette
sns.palplot(sns.color_palette('autumn'))
```

```python
#Reversing the lightness ramp in the palette
sns.palplot(sns.color_palette('autumn_r'))
```

If you want to restrict the range of colors, to create a dark palette. This comes in handy while mapping lines or points for clarity.

```python
sns.palplot(sns.color_palette('autumn_d'))
```

```python
sns.palplot(sns.light_palette("orange"))
```

#### Plotting a kdeplot with the above color palette. We use it as an object and pass it to the color map variable.

```python
plt.subplots(figsize = (10,5))
cmap = sns.light_palette("orange",as_cmap = True,reverse = True)
sns.kdeplot(pokemon_data['Attack Point'],pokemon_data['Defense Point'],cmap = cmap)
plt.xlim(-10,200)
plt.ylim(-10,250)
plt.show()
```

#### A reversed dark palette is similar in intensities to a light palette

### To interactively see the palettes available for sequential distributions of data.

```python
sns.palplot(sns.dark_palette('blue',reverse=True))
```

```python
sns.choose_colorbrewer_palette('sequential')
```

### Cubehelix Palettes
This is a sequential palette with a linear increase or decrease in brightness, along with hue variation.


```python
#Default cubehelix from matplotlib
sns.palplot(sns.color_palette('cubehelix',12))
```

```python
#Calling the default cubehelixpalette in Seaborn,
sns.palplot(sns.cubehelix_palette(12))
```

```python
#setting the size for the plot
plt.subplots(figsize=(10,5))

cmap=sns.cubehelix_palette(as_cmap=True,dark=1,light=0,reverse=True)
sns.kdeplot(pokemon_data['Attack Point'],pokemon_data['Total Point'],cmap=cmap,shade=True)
#We can set the n_levels parameter to specify number of hue levels we need
plt.xlim(-10,150)
plt.ylim(-10,800)
```

## Diverging Color Palettes
Befor we discuss this, let us first see a diverging color palette

```python
#A Color Brew Palette
sns.palplot(sns.color_palette('BrBG',9))
```

#### Custom diverging palettes
* Use the Seaborn diverging_palette() and create your own custom palette.
* Run choose_diverging_palette() from an ipython notebook for assistance
* The function uses "husl" system to create well balanced color maps


Lets view the options in diverging palettes. This time, instead of just seeing it, we shall store the object in 'h' and use this in our next plot.

```python
#A matplotlib palette
sns.palplot(sns.color_palette('summer',9))
```

```python
h=sns.choose_colorbrewer_palette('diverging')
```

We defined 'hue' as the palette 'summer'. In that case, if we plot the same graph as above with the palette name itself, we would expect the plots to be the same. Let's verify that.

```python
g=sns.PairGrid(pokemon_data,
              vars=['Attack Point','Defense Point','Speed Point'],
              hue='Total Point',
              palette='summer')
g.map_offdiag(plt.scatter)
g.map_diag(sns.kdeplot)
g.add_legend()
```

```python
### Here the color palette is changed to winter.
```

```python
g=sns.PairGrid(pokemon_data,
              vars=['Attack Point','Defense Point','Speed Point'],
              hue='Sp. Atk Point',
              palette='winter')
g.map_offdiag(plt.scatter)
g.map_diag(sns.kdeplot)
g.add_legend()
plt.show()
```
