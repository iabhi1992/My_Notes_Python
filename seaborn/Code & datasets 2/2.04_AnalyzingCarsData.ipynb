```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Here the stored dataset car_crashes is being loaded and it is analysed to find out relationship between different columns of the dataset.

```python
crashes = sns.load_dataset('car_crashes')
print(crashes.head())
```

```python
plt.subplots(figsize=(11,10))
sns.distplot(crashes['total'],
             bins=30,
             kde=False,
             rug=True)
```

Maximum frequency of total accidents occured for different states lie between 17.5 and around 20.

```python
sns.jointplot(x='alcohol',
              y='speeding',
              data=crashes,
              size=8)
plt.show()
```

From this joiont plot we can observe a linear realationship between alcohol consumption and speeding i.e higher the alcohol consumption higher the speeding.

```python
sns.jointplot(x='no_previous',
              y='speeding',
              data=crashes,
              size=8)
plt.show()
```

We can observe from this plot that ,the speeding has relation with the number of previous accidents. Most of the data points indicates that the people who speeds are having previous accident records.

```python
sns.jointplot(x='alcohol',
              y='no_previous',
              data=crashes,
              size=8)
plt.show()
```

The obvious outcome of this plot is that higher alcohol consumption is lineraly realted to previous number of accidents.

```python
sns.pairplot(crashes,
            size=5,
             aspect=0.9,
            x_vars=['ins_premium','speeding','not_distracted','no_previous'],
            y_vars=['alcohol'],
            kind='reg')
plt.show()
```
